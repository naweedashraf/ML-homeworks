<h2>Question 1</h2>
<p>Given random variable X distributed noramlly($ X\sim N(m,s^2) $) calculate Y = aX + b distribution.</p>

<h2>Answer:</h2>
We have $ m = E[X] $ and $s^2 = Var(X)$. We have to calculate $ E[Y]$ and $Var(Y)$
<br>
$E[Y] = E[aX + b] = E[aX] + b = aE[x] + b = $<strong> am + b</strong>
<br>
$Var(Y) = Var(aX + b) = Var(aX) = a^2Var(X) = a^2s^2$
<br><br>
so we have: <strong> $$Y\sim N(am + b,a^2s^2)$$

<h2>Question 2</h2>
Write a python program that produces 5 random variables considering $m = E[X] = 2$ and $s = Var(x) = 1.5$ then calculate E[X] and Var(X) of random variables and compare them to m and s.<br>
run program for N = 5, 10, 50, 100, 500, 1000 and plot E[x] and Var(x) based on the number of samples.<br>
for N = 10, 50 draw their histogram.

<h2>Answer</h2>
a)

```python
mu = 2
sigma = 1.5
n = 5
x = np.random.normal(mu, sigma, n)
print ("mean:\t\t", np.mean(x), "\tm: ", mu)
print ("variance:\t", np.std(x, ddof=1),"\ts: ",sigma)
```

b)

```python
y = []
number = []
#create samples
for i in range(6):
    #justify n
    n = pow(10,int(i/2)+1)
    if(i%2 == 0):
        n = int(n/2)
    number.append(n)
    #add samples to array
    y.append(np.random.normal(mu, sigma, n))
fig = plt.figure()
for i in range(6):
    print("number of samples: ", number[i])
    print("mean:\t\t", np.mean(y[i]), "\tm: ", mu)
    print ("variance:\t", np.std(y[i], ddof=1),"\ts: ",sigma,"\n")
    plt.scatter(number[i], np.mean(y[i]), s = 50, c = "blue")

plt.xlabel('number of samples') 
plt.ylabel('mean') 
plt.title("mean based on number of samples")
plt.show()

for i in range(6):
    plt.scatter(number[i], np.std(y[i]), s = 50, c = "blue")

plt.xlabel('number of samples') 
plt.ylabel('variance') 
plt.title("variance based on number of samples")
plt.show()
```

<h2>Question 4</h2>
write a program that produces 20 2D data X=[X1,X2] with independent distributions $ X_{1}\sim N(m_{1},s_{1}^2) $ and $ X_{2}\sim N(m_{2},s_{2}^2) $ draw them in 2 dimensional space.<br>
write 2 dimensional distribution function of X.<br>
$m_{1} = 0.5, m_{2} = 0.5$<br><br>
$s_{1}^2 = 0.1, s_{2}^2 = 0.1$<br><br>
$\Sigma =
\begin{pmatrix}
0.1 & ?\\
? & 0.1
\end{pmatrix}$

<h2>Answer:</h2>
$\Sigma =
\begin{pmatrix}
V(x_{1}) & COV(x_{1},x_{2})\\
COV(x_{2},x_{1}) & V(x_{2})
\end{pmatrix}$<br><br>
$COV(x_{i},x_{j}) = E[(Y_{i} - E[Y_{i}])(Y_{j} - E[Y_{j} ])] = E[Y_{i}Y_{j} ] - E[Y_{i}]E[Y_{j} ]$<br><br>
in 2D $\Sigma =
\begin{pmatrix}
s_{1}^2 & s_{12}^2\\
s_{21}^2 & s_{2}^2
\end{pmatrix} = 
\begin{pmatrix}
s_{1}^2 & \rho s_{1}s_{2}\\
\rho s_{1}s_{2} & s_{2}^2
\end{pmatrix}$
where $\rho$ is correlation coefficient and equals to $\frac{Cov(x_{1},x_{2})}{\sqrt{V(x_{1})V(x_{2})}} = \frac{s_{12}^2}{s_{1}s_{2}}$<br>
since x1 and x2 are independent then $\rho = 0$ and 
$\Sigma =
\begin{pmatrix}
0.1 & 0\\
0 & 0.1
\end{pmatrix}
$
<br><br>
finally distribution function of x is:
$$
N(X|\mu,\Sigma) = \frac{1}{2\pi|\Sigma|^{1/2}}exp[-\frac{1}{2}(X-\mu)^T\Sigma^{-1}(X-\mu)]
$$<br><br>

$$
where: \mu = \begin{pmatrix}
m_{1}\\
m_{2}
\end{pmatrix}
= \begin{pmatrix}
0.5\\
0.5
\end{pmatrix} and
\Sigma =
\begin{pmatrix}
0.1 & 0\\
0 & 0.1
\end{pmatrix}
$$

```python
mu1, mu2 = 0.5, 0.5
sigma1, sigma2 = 0.1, 0.1
n = 20
x1 = np.random.normal(mu1, sigma1, n)
x2 = np.random.normal(mu2, sigma2, n)
x=[x1, x2]
plt.scatter(x1,x2, s = 50, c = "blue")
plt.ylim(-3, 2)
```
